```python

# 필요한 패키지
import warnings
warnings.simplefilter('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('font', family='malgun gothic')
plt.rc('axes', unicode_minus=False)
```

## 20~22년 구별 위험요인인식정도

```python
danger_df = pd.read_csv('./서울시_자치구별_현대사회위험요인정도(범죄).csv', encoding='utf-8')
danger_df = danger_df.set_index('구분별')  # 인덱스 설정
danger_df = danger_df.rename_axis(mapper='자치구',axis=0)  # 인덱스명 설정
danger_df = danger_df.rename_axis(mapper='연도별 위험요인인식',axis=1)  # 컬럼명 설정
danger_df = danger_df.sort_index()
danger_df.info()

danger_df.head()
```

## 연도별 위험인식정도 그래프

```python

colors = ['orange', 'blue', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'cyan', 'magenta',
          'lime', 'olive', 'teal', 'navy', 'maroon', 'sienna', 'indigo', 'coral', 'salmon',
          'peru', 'gold', 'skyblue', 'darkorange', 'darkgreen', 'darkred']
color_index = 0
# 그래프 사이즈 설정
plt.figure(figsize=(10, 8))

# DF의 인덱스와 값으로 라인그래프
for index, row in danger_df.iterrows():
    plt.plot(danger_df.columns, row, marker='o', label=index, color=colors[color_index])
    color_index += 1

plt.title('연도별 위험요인인식정도')
plt.xlabel('연도')
plt.ylabel('위험요인인식정도(10점 만점)')
plt.legend(loc='best', ncol=2)
plt.grid(True)

plt.show()
```

```python
target_danger_df = danger_df.loc[['송파구', '강남구', '강서구', '종로구', '도봉구']]

colors = ['orange', 'blue', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'cyan', 'magenta',
          'lime', 'olive', 'teal', 'navy', 'maroon', 'sienna', 'indigo', 'coral', 'salmon',
          'peru', 'gold', 'skyblue', 'darkorange', 'darkgreen', 'darkred']
color_index = 0
# 그래프 사이즈 설정
plt.figure(figsize=(10, 8))

# DF의 인덱스와 값으로 라인그래프
for index, row in target_danger_df.iterrows():
    plt.plot(target_danger_df.columns, row, marker='o', label=index, color=colors[color_index])
    color_index += 1

plt.title('연도별 위험요인인식정도')
plt.xlabel('연도')
plt.ylabel('위험요인인식정도(10점 만점)')
plt.legend(loc='best', ncol=2)
plt.grid(True)

plt.show()
```
