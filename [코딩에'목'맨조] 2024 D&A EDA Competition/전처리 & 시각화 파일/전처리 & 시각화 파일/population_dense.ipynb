```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('font', family='malgun gothic')
plt.rc('axes', unicode_minus=False)
```

## 15~23년 구별 인구밀도

```python
pop_density_df = pd.read_excel('서울시_자치구별_인구밀도.xlsx')
pop_density_df = pop_density_df.drop(labels=pop_density_df.index[:2])
pop_density_df = pop_density_df.drop(labels=pop_density_df.columns[0], axis=1)
pop_density_df = pop_density_df.set_index('동별(2)')
pop_density_df = pop_density_df.rename_axis(mapper='자치구',axis=0)
pop_density_df = pop_density_df.rename_axis(mapper='인구밀도',axis=1)
pop_density_df = pop_density_df.sort_index()
pop_density_df.head()
```

## 연도별 인구밀도

```python
colors = ['orange', 'blue', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'cyan', 'magenta',
          'lime', 'olive', 'teal', 'navy', 'maroon', 'sienna', 'indigo', 'coral', 'salmon',
          'peru', 'gold', 'skyblue', 'darkorange', 'darkgreen', 'darkred']
color_index = 0

plt.figure(figsize=(10, 8))
for index, row in pop_density_df.iterrows():
    plt.plot(pop_density_df.columns, row, marker='o', label=index, color=colors[color_index])
    color_index += 1

plt.title('서울시 자치구별 인구밀도')
plt.xlabel('연도')
plt.ylabel('인구밀도')
plt.legend(loc='best', ncol=2)
plt.grid(True)

plt.show()
```

## 5개구 연도별 인구밀도

```python
target_pop_density_df = pop_density_df.loc[['송파구', '강남구', '강서구', '종로구', '도봉구']]
colors = ['orange', 'blue', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'cyan', 'magenta',
          'lime', 'olive', 'teal', 'navy', 'maroon', 'sienna', 'indigo', 'coral', 'salmon',
          'peru', 'gold', 'skyblue', 'darkorange', 'darkgreen', 'darkred']
color_index = 0
# 그래프 사이즈 설정
plt.figure(figsize=(10, 8))
for index, row in target_pop_density_df.iterrows():
    plt.plot(target_pop_density_df.columns, row, marker='o', label=index, color=colors[color_index])
    color_index += 1

plt.title('서울시 연도별 인구밀도')
plt.xlabel('연도')
plt.ylabel('인구밀도')
plt.legend(loc='best', ncol=2)
plt.grid(True)

plt.show()
```
